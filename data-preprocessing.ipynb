## generate negative examples with random pairing

```python
# get sequences from unpaired datasets
human_unpaired_heavy_df = pd.read_csv(human_unpaired_heavy_path)[['seq']].dropna()
human_unpaired_heavy_df.columns = ['heavy']
human_unpaired_light_df = pd.read_csv(human_unpaired_light_path)[['seq']].dropna()
human_unpaired_light_df.columns = ['light']
BALB_unpaired_heavy_df = pd.read_csv(BALB_unpaired_heavy_path)[['seq']].dropna()
BALB_unpaired_heavy_df.columns = ['heavy']
BALB_unpaired_light_df = pd.read_csv(BALB_unpaired_light_path)[['seq']].dropna()
BALB_unpaired_light_df.columns = ['light']
C57BL_unpaired_heavy_df = pd.read_csv(C57BL_unpaired_heavy_path)[['seq']].dropna()
C57BL_unpaired_heavy_df.columns = ['heavy']
C57BL_unpaired_light_df = pd.read_csv(C57BL_unpaired_light_path)[['seq']].dropna()
C57BL_unpaired_light_df.columns = ['light']
rhesus_unpaired_heavy_df = pd.read_csv(rhesus_unpaired_heavy_path)[['seq']].dropna()
rhesus_unpaired_heavy_df.columns = ['heavy']
rhesus_unpaired_light_df = pd.read_csv(rhesus_unpaired_light_path)[['seq']].dropna()
rhesus_unpaired_light_df.columns = ['light']
rabbit_unpaired_heavy_df = pd.read_csv(rabbit_unpaired_heavy_path)[['seq']].dropna()
rabbit_unpaired_heavy_df.columns = ['heavy']
rat_unpaired_heavy_df = pd.read_csv(rat_unpaired_heavy_path)[['seq']].dropna()
rat_unpaired_heavy_df.columns = ['heavy']
camel_unpaired_heavy_df = pd.read_csv(camel_unpaired_heavy_path)[['seq']].dropna()
camel_unpaired_heavy_df.columns = ['heavy']
```

```python
def integer_split(num, div):
    split_list = [num // div + (1 if x < num % div else 0)  for x in range (div)]
    return(split_list)

def generate_negatives(paired, unpaired_heavy, unpaired_light, light_dfs, heavy_dfs):
    halfway_cutoff = math.floor(paired.shape[0]/2)
    
    # generate split lists
    split_list_light = integer_split(paired.shape[0] - halfway_cutoff, len(light_dfs))
    split_list_heavy = integer_split(paired.shape[0] - halfway_cutoff, len(heavy_dfs))
    
    paired_heavy = paired[['heavy']]
    paired_light = paired[['light']]
    
    # make light stack
    light_stack = unpaired_light[0:halfway_cutoff]
    for i,light_df in enumerate(light_dfs):
        split = split_list_light[i]
        light_stack = pd.concat([light_stack, light_df[0:split]], axis=0)
    
    # make heavy stack
    heavy_stack = unpaired_heavy[0:halfway_cutoff]
    for i,heavy_df in enumerate(heavy_dfs):
        split = split_list_heavy[i]
        heavy_stack = pd.concat([heavy_stack, heavy_df[0:split]], axis=0)
        
    # extend light and heavy stacks
    heavy_stack = pd.concat([paired_heavy, heavy_stack]).reset_index(drop=True)
    light_stack = pd.concat([light_stack, paired_light]).reset_index(drop=True)
    
    # merge two stacks horizontally
    negatives_df = pd.concat([heavy_stack, light_stack], axis=1)
    negatives_df['paired'] = 0
    
    return negatives_df
```

```python
human_light_dfs = [BALB_unpaired_light_df, C57BL_unpaired_light_df, rhesus_unpaired_light_df]
BALB_light_dfs = [human_unpaired_light_df, C57BL_unpaired_light_df, rhesus_unpaired_light_df]
C57BL_light_dfs = [human_unpaired_light_df, BALB_unpaired_light_df, rhesus_unpaired_light_df]

human_heavy_dfs = [BALB_unpaired_heavy_df, C57BL_unpaired_heavy_df, rhesus_unpaired_heavy_df,
                   rabbit_unpaired_heavy_df, rat_unpaired_heavy_df, camel_unpaired_heavy_df]
BALB_heavy_dfs = [human_unpaired_heavy_df, C57BL_unpaired_heavy_df, rhesus_unpaired_heavy_df,
                   rabbit_unpaired_heavy_df, rat_unpaired_heavy_df, camel_unpaired_heavy_df]
C57BL_heavy_dfs = [human_unpaired_heavy_df, BALB_unpaired_heavy_df, rhesus_unpaired_heavy_df,
                   rabbit_unpaired_heavy_df, rat_unpaired_heavy_df, camel_unpaired_heavy_df]
```

```python
human_negatives = generate_negatives(human_paired_df, human_unpaired_heavy_df, human_unpaired_light_df, human_light_dfs, human_heavy_dfs)
BALB_negatives = generate_negatives(BALB_paired_df, BALB_unpaired_heavy_df, BALB_unpaired_light_df, BALB_light_dfs, BALB_heavy_dfs)
C57BL_negatives = generate_negatives(C57BL_paired_df, C57BL_unpaired_heavy_df, C57BL_unpaired_light_df, C57BL_light_dfs, C57BL_heavy_dfs)

full_unpaired_df = pd.concat([human_negatives, BALB_negatives, C57BL_negatives])
final_dataset = pd.concat([full_paired_df, full_unpaired_df]).dropna()
```
